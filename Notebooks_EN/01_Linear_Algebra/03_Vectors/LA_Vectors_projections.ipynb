# Vectors - Projections and Divisions


## Projecting a Vector onto Another Vector

Projecting vector $\mathbf{a}$ onto axis $\mathbf{b}$ involves finding the component of vector $\mathbf{a}$ along the direction of $\mathbf{b}$.

Projection of $\mathbf{a}$ onto $\mathbf{b}$:

$$
\text{Proj}_{\mathbf{b}} \mathbf{a} = \frac{\mathbf{a} \cdot \mathbf{b}}{|\mathbf{b}|^2} \mathbf{b}
$$


## Projecting a Vector onto Axes

Projecting vector $\mathbf{a}$ onto an axis involves finding the component of vector $\mathbf{a}$ along that axis.

Projection of $\mathbf{a}$ onto the $OX$ axis:

$$
\text{Proj}_{OX} \mathbf{a} = \frac{\mathbf{a} \cdot \mathbf{i}}{|\mathbf{i}|^2} \mathbf{i} = \frac{\mathbf{a} \cdot \mathbf{i}}{1} \mathbf{i} = (\mathbf{a} \cdot \mathbf{i}) \mathbf{i}
$$

Projection of $\mathbf{a}$ onto the $OY$ axis:

$$
\text{Proj}_{OY} \mathbf{a} = \frac{\mathbf{a} \cdot \mathbf{j}}{|\mathbf{j}|^2} \mathbf{j} = \frac{\mathbf{a} \cdot \mathbf{j}}{1} \mathbf{j} = (\mathbf{a} \cdot \mathbf{j}) \mathbf{j}
$$

Recall that $\mathbf{i} = (1, 0)$ and $\mathbf{j} = (0, 1)$ are unit vectors along the $OX$ and $OY$ axes.


## Dividing a Line Segment in a Given Ratio

A line segment is defined by two points in space.

Line segments and vectors are closely related because a line segment connecting two points can be treated as a vector directed from one point to another.

If point \( P \) divides the line segment connecting points $A(x_1, y_1)$ and $B(x_2, y_2)$ in the ratio $m:n$, then the coordinates of point $P$ can be found using the formula:

$$
P = \left(\frac{n \times x_2 + m \times x_1}{m+n}, \frac{n \times y_2 + m \times y_1}{m+n}\right)
$$

**Example:**

Given a line segment connecting points $A(1, 2)$ and $B(3, 8)$, find point $P$ that divides the segment in the ratio 2:3.

Coordinates of point $P$:
$$
P = \left(\frac{2 \times 3 + 3 \times 1}{2+3}, \frac{2 \times 8 + 3 \times 2}{2+3}\right) = \left(\frac{9}{5}, \frac{22}{5}\right)
$$

```python
import matplotlib.pyplot as plt
import math

# Coordinates of points A and B
A = (1, 2)
B = (3, 8)

# Ratio m:n
m, n = 2, 3

# Calculate coordinates of point P that divides the segment in the ratio 2:3
P_x = (n * A[0] + m * B[0]) / (m + n)
P_y = (n * A[1] + m * B[1]) / (m + n)
P = (P_x, P_y)

# Plotting the graph
plt.figure(figsize=(6, 6))

# Plotting line segment AB
plt.plot([A[0], B[0]], [A[1], B[1]], 'k--', label='Line Segment AB')

# Marking points A, B, and P
plt.scatter(*A, color='blue', label='A(1, 2)')
plt.scatter(*B, color='red', label='B(3, 8)')
plt.scatter(*P, color='green', label=f'P({P[0]:.2f}, {P[1]:.2f})')

# Point labels
plt.text(A[0]-0.2, A[1], 'A(1, 2)', fontsize=12, color='blue')
plt.text(B[0]+0.1, B[1], 'B(3, 8)', fontsize=12, color='red')
plt.text(P[0]+0.1, P[1], f'P({P[0]:.2f}, {P[1]:.2f})', fontsize=12, color='green')

# Adding grid and axis labels
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Dividing a Line Segment in the Ratio 2:3')
plt.legend()

# Displaying the graph
plt.show()
```

## Student Exercises

- Divide the line segment connecting points $A(-1, 2)$ and $B(3, -2)$ in the ratio $1:3$. Illustrate the result on a graph.

- Project vector $\mathbf{a} = (3, 4)$ onto the $OX$ and $OY$ axes. Illustrate the result on a graph.

- Project vector $\mathbf{a} = (2,3)$ onto vector $\mathbf{b} = (1, 1)$. Illustrate the result on a graph.

- Project vector $\mathbf{b} = (1, 1)$ onto vector $\mathbf{a} = (2, 3)$. Illustrate the result on a graph.

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to plot vectors
def plot_vector(v, origin=[0, 0], color='b', label=''):
    plt.quiver(*origin, *v, color=color, scale=1, scale_units='xy', angles='xy', label=label)

# 1. Divide the line segment connecting points A(-1, 2) and B(3, -2) in the ratio 1:3
A = np.array([-1, 2])
B = np.array([3, -2])

def divide_line_segment(A, B, ratio):
    m, n = ratio
    # Using the section formula: P = (n*A + m*B) / (m+n)
    P = (n * A + m * B) / (m + n)
    return P

# Divide in the ratio 1:3
P = divide_line_segment(A, B, (1, 3))

print(f"Point dividing the segment AB in ratio 1:3: {P}")

# Plotting the points and the division
plt.figure()
plt.scatter(*A, color='r', label='A(-1, 2)')
plt.scatter(*B, color='g', label='B(3, -2)')
plt.scatter(*P, color='b', label=f'P({P[0]:.1f}, {P[1]:.1f})')
plt.plot([A[0], B[0]], [A[1], B[1]], 'k--', label='Segment AB')

plt.legend()
plt.grid(True)
plt.title("Line Segment Division in Ratio 1:3")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()

# 2. Project vector a = (3, 4) onto the OX and OY axes. Illustrate the result on a graph.
a = np.array([3, 4])

# Projection onto OX axis (x-component) and OY axis (y-component)
proj_OX = np.array([a[0], 0])
proj_OY = np.array([0, a[1]])

print(f"Projection of a onto OX: {proj_OX}")
print(f"Projection of a onto OY: {proj_OY}")

# Plot the original vector and its projections
plt.figure()
plot_vector(a, label='a = (3, 4)')
plot_vector(proj_OX, color='r', label='Projection onto OX')
plot_vector(proj_OY, color='g', label='Projection onto OY')

plt.xlim(0, 5)
plt.ylim(0, 5)
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.title('Projection of a onto OX and OY')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()

# 3. Project vector a = (2, 3) onto vector b = (1, 1). Illustrate the result on a graph.
a = np.array([2, 3])
b = np.array([1, 1])

def project_vector(u, v):
    # Projection formula: proj_v(u) = (u . v / v . v) * v
    return (np.dot(u, v) / np.dot(v, v)) * v

proj_ab = project_vector(a, b)

print(f"Projection of a onto b: {proj_ab}")

# Plot vectors a, b, and the projection of a onto b
plt.figure()
plot_vector(a, label='a = (2, 3)')
plot_vector(b, color='g', label='b = (1, 1)')
plot_vector(proj_ab, color='r', label='Projection of a onto b')

plt.xlim(0, 3)
plt.ylim(0, 3)
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.title('Projection of a onto b')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()

# 4. Project vector b = (1, 1) onto vector a = (2, 3). Illustrate the result on a graph.
proj_ba = project_vector(b, a)

print(f"Projection of b onto a: {proj_ba}")

# Plot vectors b, a, and the projection of b onto a
plt.figure()
plot_vector(b, label='b = (1, 1)')
plot_vector(a, color='g', label='a = (2, 3)')
plot_vector(proj_ba, color='r', label='Projection of b onto a')

plt.xlim(0, 3)
plt.ylim(0, 3)
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.title('Projection of b onto a')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
